```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dane = pd.DataFrame({'Miesiąc': ["Styczeń","Luty","Marzec","Kwiecień","Maj","Czerwiec","Lipiec","Sierpień","Wrzesień","Październik","Listopad","Grudzień"],
                     'SmartTV': [159,187,245,151,181,160,142,217,152,143,157,175],
                     'TV': [159,187,246,151,181,150,134,149,131,105,129,142]})
```

### Wykres domyślny

```python
dane.plot(figsize=[9, 5])
plt.title("Sprzedaż telewizorów w roku 2016")
plt.xlabel("Miesiące")
plt.xticks(dane.index, dane.Miesiąc);
```

### Poprawienie wykresu zgodnie z regułami

##### 1. Usuń elementy nie wnoszące nic do zrozumienia prezentowanego problemu. Zastosuj:

- Prawo domykania
- FactorInk

```python
dane.plot(figsize=[9, 5])
plt.title("Sprzedaż telewizorów w roku 2016")
plt.xlabel("Miesiące")
plt.xticks(dane.index, dane.Miesiąc);

# Usunięcie ramki z wykresu
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.show()
```

##### 2. Użyj najkrótszych jak to tylko możliwe etykiet danych, pozwól odbiorcy czytać "naturalnie". Zastosuj:

- Prawo kontynuacji



```python
# Mapowanie skrótów miesięcy
skroty_miesiecy = {
    "Styczeń": "Sty",
    "Luty": "Lut",
    "Marzec": "Mar",
    "Kwiecień": "Kwi",
    "Maj": "Maj",
    "Czerwiec": "Cze",
    "Lipiec": "Lip",
    "Sierpień": "Sie",
    "Wrzesień": "Wrz",
    "Październik": "Paź",
    "Listopad": "Lis",
    "Grudzień": "Gru"
}
dane['Miesiąc'] = dane['Miesiąc'].map(skroty_miesiecy)

dane.plot(figsize=[9, 5])
plt.title("Sprzedaż telewizorów w roku 2016")
plt.xlabel("Miesiące")
plt.xticks(dane.index, dane.Miesiąc);


# Ustawienie etykiet na osi X
plt.xticks(range(len(dane)), dane['Miesiąc'])

# Usunięcie ramki z wykresu
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.show()
```

##### 3. Usuń elementy rozpraszające (legendę). Zastosuj:

- Prawo bliskości
- Prawo podobieństwa

```python
dane.plot(figsize=[9, 5])
plt.title("Sprzedaż telewizorów w roku 2016")
plt.xlabel("Miesiące")
plt.xticks(dane.index, dane.Miesiąc);

# Ustawienie etykiet na osi X
plt.xticks(range(len(dane)), dane['Miesiąc'])

# Usunięcie ramki z wykresu
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.legend().remove()

plt.text(11, 146, 'TV', color='orange', fontsize=10)
plt.text(11, 168, 'SmartTV', color='blue', fontsize=10)

plt.show()
```

##### 4. Pomóż czytelnikowi zwrócić uwagę na istotne elementy. Pokaż mu to co chcesz żeby zobaczył.

```python
kolory = ['blue', 'orange']
dane.plot(figsize=[9, 5], color = kolory)
plt.title("Sprzedaż telewizorów w roku 2016")
plt.xlabel("Miesiące")
plt.xticks(dane.index, dane.Miesiąc);
# Ustawienie etykiet na osi X
plt.xticks(range(len(dane)), dane['Miesiąc'])

# Usunięcie ramki z wykresu
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.legend().remove()


plt.text(11.2, 146, 'TV', color=kolory[1], fontsize=10)
plt.text(11.2, 168, 'SmartTV', color=kolory[0], fontsize=10)


yticks = (0, 50, 100, 150, 200, 250, 300)
plt.yticks(yticks)
plt.show()
```

```python
kolory = ['blue', 'orange']
dane.plot(figsize=[9, 5], color = kolory)
plt.title("Sprzedaż telewizorów w roku 2016")
plt.xticks(dane.index, dane.Miesiąc);
# Ustawienie etykiet na osi X
plt.xticks(range(len(dane)), dane['Miesiąc'])

# Usunięcie ramki z wykresu
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.legend().remove()

plt.text(11.2, 135, 'TV', color=kolory[1], fontsize=11,  fontweight='bold')
plt.text(11.2, 179, 'SmartTV', color=kolory[0], fontsize=11,  fontweight='bold')
yticks = (0, 50, 100, 150, 200, 250, 300)
plt.yticks(yticks)

maj_index = dane[dane['Miesiąc'] == 'Maj'].index[0]
gru_index = dane[dane['Miesiąc'] == 'Gru'].index[0]
plt.axvspan(maj_index - 0.25, maj_index + 0.25, color='lightgrey', alpha=0.5)
plt.text(maj_index, 270, 'Netflix', horizontalalignment='center', fontsize=12, color='black',  fontweight='bold')

roznica_grudzien = dane.loc[dane['Miesiąc'] == 'Gru', 'TV'].values[0] - dane.loc[dane['Miesiąc'] == 'Gru', 'SmartTV'].values[0]
plt.text(gru_index + 0.5, 155, roznica_grudzien, horizontalalignment='center', fontsize=11, color='red')

plt.show()
```
